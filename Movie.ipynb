```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer # It takes some key words and covert them in to the numbers.
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
df=pd.read_csv("/content/titles.csv",engine='python')
```

```python
df['sno'] = range(0, len(df))
```

```python
df.set_index('sno', inplace=True)
```

```python
df.head()
```

```python
selected_column=['title','description','type','genres']
for i in selected_column:
  df[i]=df[i].fillna('')
```

```python

combine_col=df[selected_column[0]]+" "+df[selected_column[1]]+" "+df[selected_column[2]]
```

```python
combine_col
```

```python
fe=TfidfVectorizer(min_df=1, stop_words='english',lowercase=True)
```

```python
extracted_col=fe.fit_transform(combine_col)
```

```python
print(extracted_col)
```

```python
title=df['title'].tolist()
```

```python
similarity=cosine_similarity(extracted_col)
```

```python
similarity.shape
```

```python
ty=input("enter movie name: ")
```

```python
similar_movie=difflib.get_close_matches(ty,title)
```

```python
similar_movie
```

```python
index=df[df.title==similar_movie[0]]['sno'].values[0]
index
```

```python
similar_movies=list(enumerate(similarity[index]))
```

```python
print(similar_movies)
```

```python
similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
```

```python
print(similar_movies)
```

```python
for i in range(20):
  print(title[similar_movies[i][0]])
```

```python
ty=input("enter movie name: ")
similar_movie=difflib.get_close_matches(ty,title)
index=df[df.title==similar_movie[0]]['sno'].values[0]
similar_movies=list(enumerate(similarity[index]))
similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
for i in range(20):
  print(title[similar_movies[i][0]])
```
